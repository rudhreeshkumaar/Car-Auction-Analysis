```python
%matplotlib inline
import math
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
import plotly.express as px
import plotly.graph_objects as go




from collections import defaultdict
from scipy.stats import pearsonr
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
df = pd.read_csv("D:/DM_Project_Dataset/training_cleaned.csv")
df.head()
```

```python
df_clus = df.drop(columns=['RefId',
        'PurchDate',
        'IsBadBuy',
        'Auction',
        'Make',
        'Model',
        'WheelTypeID',
        'Nationality',
        'Size',
        'Color',
        'Transmission',
        'BYRNO',
        'TopThreeAmericanName',
        'SubModel',
        'VehicleAge',
        'IsOnlineSale',
        'Region',
        'IsBase',])
df_clus.dropna(inplace=True)
len(df_clus)
df_clus.head()
```

# K-Means with Z-Score

```python
scaler = StandardScaler()
z_df_clus = scaler.fit_transform(df_clus.values)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
kmeans = KMeans(n_clusters=5, n_init=50, max_iter=500)
kmeans.fit(z_df_clus)
```

```python
plt.scatter(df_clus['AuctionAVG'], df_clus['WarrantyCost'], 
            c=kmeans.labels_, s=60)
```

# K-Means with MinMax

```python
scaler = MinMaxScaler()
minmax_df_clus = scaler.fit_transform(df_clus.values)
```

```python
kmeans_2 = KMeans(n_clusters=5, n_init=50, max_iter=500)
kmeans_2.fit(minmax_df_clus)
```

```python
plt.scatter(df_clus['WarrantyCost'], df_clus['VehBCost'], 
            c=kmeans_2.labels_, s=60)
```
